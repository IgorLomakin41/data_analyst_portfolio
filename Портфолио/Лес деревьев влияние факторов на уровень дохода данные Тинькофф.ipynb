```python
#Оставим данные где есть ответ на вопрос об уровне дохода
df=df.drop(df[df['k85']=='ОТКАЗ ОТ ОТВЕТА'].index)
df=df.drop(df[df['k85']=='НЕТ ОТВЕТА'].index)
df=df.drop(df[df['k85']=='ЗАТРУДНЯЮСЬ ОТВЕТИТЬ'].index)
df['k85'].value_counts()
```

```python
#Выделим числовой столбец с уровнем дохода по возрастанию по показателям. 
df['levels_income']=df['k85'].map({'ДЕНЕГ НЕ ХВАТАЕТ ДАЖЕ НА ПИТАНИЕ':0, 'НА ПИТАНИЕ ДЕНЕГ ХВАТАЕТ, НО НЕ ХВАТАЕТ НА ПОКУПКУ ОДЕЖДЫ И ОБУВИ':1,
                                      'НА ПОКУПКУ ОДЕЖДЫ И ОБУВИ ДЕНЕГ ХВАТАЕТ, НО НЕ ХВАТАЕТ НА ПОКУПКУ КРУПНОЙ БЫТОВОЙ ТЕХНИКИ':2,
                                      'ДЕНЕГ ВПОЛНЕ ХВАТАЕТ НА ПОКУПКУ КРУПНОЙ БЫТОВОЙ ТЕХНИКИ, НО МЫ НЕ МОЖЕМ КУПИТЬ АВТОМАШИНУ':3,
                                      'ДЕНЕГ ХВАТАЕТ НА ВСЁ, КРОМЕ ТАКИХ ПРИОБРЕТЕНИЙ КАК КВАРТИРА, ДОМ':4,
                                      'МАТЕРИАЛЬНЫХ ЗАТРУДНЕНИЙ НЕ ИСПЫТЫВАЕМ, ПРИ НЕОБХОДИМОСТИ МОГЛИ БЫ КУПИТЬ КВАРТИРУ, ДОМ':5})
df['levels_income'].value_counts()
```

```python
#Превратим категориальные столбцы в бинарные
X=df.drop(['levels_income','k85'], axis=1)
y=df['levels_income']
X=pd.get_dummies(X)
```

```python
#Рассчитаем лучшие параметры леса деревьев 
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
parametrs={'n_estimators':[200,300,400], 'max_depth':[5,7,10,12], 'random_state':[0,42]}
```

```python
#from sklearn.model_selection import GridSearchCV
#gsv=GridSearchCV(clf, parametrs, cv=5)
#gsv.fit(X,y)
#gsv.best_params_
```

```python
#Обучим модель на выбранных параметрах
model = RandomForestClassifier(max_depth=12, n_estimators=300, random_state=0)
model.fit(X, y)
```

```python
model.feature_importances_
```

```python
#Выберем важные параметры и отсортируем их по убыванию
feature_importances=model.feature_importances_
feature_importances_df=pd.DataFrame({'features':list(X),'feature_importances': feature_importances})

feature_importances_df=feature_importances_df.sort_values('feature_importances',ascending=False)
feature_importances_df.head(10)
```

```python
#Уберем лишнее
X=X.drop(['X','id_i', 'Unnamed: 0' ], axis=1)
model = RandomForestClassifier(max_depth=12, n_estimators=300, random_state=0)
model.fit(X, y)
```

```python
feature_importances=model.feature_importances_
feature_importances_df=pd.DataFrame({'features':list(X),'feature_importances': feature_importances*100})

feature_importances_df=feature_importances_df.sort_values('feature_importances',ascending=False)
feature_importances_df.head(25)
```
